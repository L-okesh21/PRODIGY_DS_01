```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Generate simulated population data
sample_size = 1000
data = {
    'Age': np.random.randint(0, 90, sample_size),  # ages from 0 to 89
    'Gender': np.random.choice(['Male', 'Female'], sample_size)
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution of Sample Population')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Gender Distribution in Sample Population')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
